```python
#import all useful libs
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#import our data
data=pd.read_csv('iris.csv')
print(data.head())
```

```python
print(data['variety'].unique)
```

```python
# Choose subset of your attributes:
df=data.iloc[:,:2]
#or
df=data[['sepal.length','sepal.width']]
```

```python
#Find mean of the data
centroid =df.mean()
print(centroid)
```

```python
plt.scatter(df.iloc[:,0], df.iloc[:,1])
plt.grid()
plt.xlabel('Sepal Length')
plt.scatter(centroid[0],centroid[1],c='red')
```

```python
#center your data
df_center=df-centroid 
plt.scatter(df_center.iloc[:,0],df_center.iloc[:,1])
plt.grid()
```

```python
line=np.array([-2.15, 2.75])
line_unit= line/np.linalg.norm(line,2)
```

```python
#project points to line (dot product)
proj=np.dot(centroid,line_unit)*line_unit
```

```python
datapro=np.dot(df_center, line_unit)[:,np.newaxis]*line_unit
```

```python
plt.scatter(datapro[:,0], datapro[:,1])
plt.scatter(df_center.iloc[:,0], df_center.iloc[:,1], c='red')
```
